```python
#sort
import pandas as pd
df=pd.read_csv("Bollywood.csv")
df1=df["MovieName"][df["Genre"]=="Thriller"]
df1.sort_values(ascending=False)
print(df1)

df10=df["MovieName"][df["Genre"]=="Romance"]
df10.sort_values(ascending=False)
print(df10)

df11=df["MovieName"][df["Genre"]=="Drama"]
df11.sort_values(ascending=False)
print(df11)

df12=df["MovieName"][df["Genre"]=="Action"]
df12.sort_values(ascending=False)
print(df12)

```

```python
#______3____
import pandas as pd
df=pd.read_csv("Bollywood.csv")
pd.crosstab(df["Genre"], df["ReleaseTime"])
```

```python
#_____4_____
df["Release Date"]=pd.to_datetime(df["Release Date"])
df["month"]=pd.DatetimeIndex(df["Release Date"]).month
df.month
c=df.month.value_counts()#month count
c
```

```python
#______5__________# 2 month 9 movies
df["Release Date"]=pd.to_datetime(df["Release Date"])
df[df.Budget>=25]["Release Date"].dt.month.value_counts()
```

```python
#______6______
df["ROI"]=df["BoxOfficeCollection"]-df["Budget"]/df["Budget"]
#df["ROI"]
df.sort_values(by="ROI").MovieName[0:10]
```

```python
#_______7_______
df.groupby(by="ReleaseTime").ROI.mean()
#df["ROI"].max()
```
