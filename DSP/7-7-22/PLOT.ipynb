## grid

```python
import matplotlib.pyplot as plt
import numpy as np
xpoints=np.array([1,2,3,4,5])
ypoints=np.array([3,8,1,10,8])
plt.plot(xpoints,ypoints)
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
xpoints=np.array([1,2,3,4,5])
ypoints=np.array([3,8,1,10,8])
plt.plot(xpoints,ypoints)
plt.grid(axis="x")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
xpoints=np.array([1,2,3,4,5])
ypoints=np.array([3,8,1,10,8])
plt.plot(xpoints,ypoints)
plt.grid(axis="y")
plt.show()
```

## subplot

```python
import matplotlib.pyplot as plt
import numpy as np

#plot 1:
x = np.array([3,4,5,6])
y = np.array([5,7,89,10])


plt.subplot(1, 2, 1)
plt.plot(x,y)
plt.title("hii")

#plot 2:
x = np.array([34,5,67,9])
y = np.array([45,67,89,2])


plt.subplot(1, 2, 2)
plt.plot(x,y)
plt.title("hello")

plt.suptitle("hoooo")
plt.show()
```
