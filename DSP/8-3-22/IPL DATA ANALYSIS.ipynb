## grouping and aggregating

```python
#What is the average SOLD PRICE for each age category
df.groupby(df["AGE"])["SOLD PRICE"].mean()
```

```python
df.groupby(df["AGE"])["SOLD PRICE"].mean().reset_index()
```

```python
#Average SOLD PRICE for Diﬀerent Playing Roles in Each Age Category
df.groupby(["PLAYING ROLE","AGE"])["SOLD PRICE"].mean().reset_index()
```

```python
#Compare the average sold price for diﬀerent ages and playing roles
soldprice_by_age=df.groupby("AGE")["SOLD PRICE"].mean()
soldprice_by_age_role=df.groupby(["PLAYING ROLE" ,"AGE"])["SOLD PRICE"].mean()
soldprice_comparison = soldprice_by_age_role.merge( soldprice_by_age, on = 'AGE',  how = 'outer') 
soldprice_comparision
```

merge() for combining data on common columns or indices
.join() for combining data on a key column or an index
