```python
import pandas as pd
df=pd.read_csv("choco.csv")
df
```

```python
#linear (ordered)
df['price']=df['price'].interpolate(method="linear")
df
```

```python
#limit decides how many consecutive values you have to fill
df['price']=df['price'].interpolate(method="linear",limit_direction="forward")
df
```

```python
df['price']=df['price'].interpolate(method="linear",limit_direction="backward")
df
```

```python
df['price']=df['price'].interpolate(method="linear",limit=1)
df
```
