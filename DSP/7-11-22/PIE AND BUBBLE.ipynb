```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([10,20,30,40,50])
col=["pink","red","blue","green","magenta"]
plt.pie(x,colors=col,labels=[1,2,3,4,5],startangle=360)
plt.legend(title="title")
plt.show()
```

## explode

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([10,20,30,40,50])
col=["pink","red","blue","green","magenta"]
myexplode=[0.3,0,0,0,0.9]
plt.pie(x,colors=col,labels=[1,2,3,4,5],startangle=360,explode=myexplode)
plt.legend(title="title")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([10,20,30,40,50])
col=["pink","red","blue","green","magenta"]
myexplode=[0.3,0,0,0,0.2]
plt.pie(x,colors=col,labels=[1,2,3,4,5],startangle=360,explode=myexplode,shadow=True)
plt.legend(title="title")
plt.show()
```

## subplot

```python
import matplotlib.pyplot as plt
import numpy as np
#plot1
x=np.array([1,2,3])
y=np.array([4,2,7])

plt.subplot(1,3,1)
plt.plot(x,y)
plt.title("hii")


#plot2
x=np.array([1,2,3])
y=np.array([4,2,7])

plt.subplot(1,3,2)
plt.plot(x,y)
plt.title("hello")
#plot3
x=np.array([10,2,3])
y=np.array([4,20,7])

plt.subplot(1,3,3)
plt.plot(x,y)
plt.title("namsthe")

plt.suptitle("ooooooooooooooooooooooooooooooo")
plt.show()

```

```python
#random colors
import matplotlib.pyplot as plt
import numpy as np
x=np.random.rand(40)
y=np.random.rand(40)
colors=np.random.rand(40)
plt.legend()
plt.scatter(x,y,s=2*100,c=colors)
plt.show()
```

```python
#alpha-trancperency
import matplotlib.pyplot as plt
import numpy as np
x=np.random.rand(40)
print(x)
y=np.random.rand(40)
colors=np.random.rand(40)
plt.scatter(x,y,s=2*100,c=colors,alpha=0.6)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.random.rand(40)
y=np.random.rand(40)
colors=np.random.rand(40)
plt.scatter(x,y,s=2*100,c=colors,alpha=0.9,marker="*")
plt.show()
```
