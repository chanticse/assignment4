## bar graph

```python
import matplotlib.pyplot as plt
country=["usa","india","swiz"]
gdp_per_capita=[234,56,78]
plt.title("COUNTRY VS GDP")
plt.xlabel("country")
plt.ylabel("GDP PER CAPITA")
plt.bar(country,gdp_per_capita,color="blue")
plt.show()
```

## barh

```python
import matplotlib.pyplot as plt
country=["usa","india","swiz"]
gdp_per_capita=[234,56,78]
plt.title("COUNTRY VS GDP")
plt.xlabel("country")
plt.ylabel("GDP PER CAPITA")
plt.barh(country,gdp_per_capita,color="green")
plt.show()
```
