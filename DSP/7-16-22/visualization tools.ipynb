# histogram

```python
import pandas as pd
import matplotlib.pyplot as plt
data={'name':['anusha','indira','ramesh','pujitha','laxmi','naidu'],
      'height':[60,61,63,65,61,60],
      'weight':[47,89,52,58,50,47]
}
df=pd.DataFrame(data)
df.plot(kind='hist',ec="black")

plt.show()
```

```python
df3.plot(kind='hist',x='YearsExperience',linestyle='--')

plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.title('salary')
plt.show()
```

# specialized visualization tools

# scatter plot


```python
import numpy as np
import matplotlib.pyplot as plt
discount=np.array([10,20,30,40,50])
saleInRs=np.array([40000,45000,48000,50000,100000])
plt.scatter(x=discount,y=saleInRs)
plt.title('sales vs discount')
plt.xlabel('discount offered')
plt.ylabel('saies in Rs')
plt.show()
```

# box plot

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data={'name':['anusha','indira','poojitha','ramesh','laxmi','naidu'],
      'english':[95,45,78,34,78,99],
      'maths':[67,78,45,78,99,89],
      'hindi':[67,87,45,89,45,65],
      'science':[68,43,89,56,98,67]      
}
df=pd.DataFrame(data)
df.plot(kind='box',color='red',title='Compare Resorts')
plt.title('performance analysis')
plt.xlabel('subjects')
plt.ylabel('marks')
plt.show()
```

# pie chart
